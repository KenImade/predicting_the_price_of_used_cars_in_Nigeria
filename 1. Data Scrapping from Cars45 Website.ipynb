### Generator to get the links for each of the cars

```python
# Building the car listing generator

car_urls = []

for i in range(1,89):
    url = 'https://buy.cars45.com/cars?page={}'.format(i)
    page = requests.get(url).text
    # print(page)
    soup = BeautifulSoup(page,'html5lib')
    listings = soup.find_all(class_='product_box')
    for l in listings:
        link = l.a['href']
        car_urls.append(link)
```

```python
from requests.exceptions import ConnectionError
requests_list = []
error_list = []
for c in range(0,len(car_urls)):
    new_url = car_urls[c]
    try:
        car_page = requests.get(new_url).text
    except ConnectionError:
        error_list.append(new_url)
    else:
        requests_list.append(car_page)
    print(c)
```
